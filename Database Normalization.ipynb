### Database Normalization 
<br>
is a process used to organize a database into tables and columns. The idea is that a table should be about a specific topic and that and only supporting topics included.

1st Normal Form : the information is stored in a relational table and each column contains atomic values, and there are not repeating groups of columns. 
**Atomic Values**

you can not store "two different things" in "one column".
and you cant store "one value" in "two different columns"

2nd Normal Form : the table is in first normal form and all the columns depend on the table's primary key.
**Composite keys/Partial Dependance**

Composite keys : makes more than one column for uniqueness of the row.
Partial Dependance : if three columns are needed for uniqueness, other rows aren't be dependant on all of them. --> should be dependant on ALL key columns

* customer address is only dependant on customer
* product manufacturer is only dependant on product

3rd Normal Form : the table is in second normal form and all of its columns are not transitively dependant on the primary key
**Transative Dependance**

* the column should be directly dependant to the primary key, not through another column
* in here address depends on product manufacturer and not the product itself


4th Normal Form : the table is in third normal form and does not contain two or more independant multi-valued facts about an entity
**Multi-valued Facts**

* we can't have two rows with same primary key in a table

in third table we dont have one specific primary key, so primary key is both columns together
